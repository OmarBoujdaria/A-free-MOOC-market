# Preprocessing Functions

```python
def url_to_text(text):
    return text.replace('/',' ').replace('-',' ').replace('_',' ')
```

```python
def lemmatize_stemming(text):
    stemmer = SnowballStemmer('english')
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))
```

```python
# Loading Negative Keywords
import pickle as pkl

NEGATIVE_KEYWORDS = list(map(lemmatize_stemming,pkl.load(open("Neg_keywords.pkl","rb"))))
```

```python

('; ').join(NEGATIVE_KEYWORDS)
```

```python
def erase_punctuation(text):
    punctuation = ',.;:()?!/"-'
    
    for char in punctuation:
        text = text.replace(char,'')

    return text
```

```python
def preprocess2(text):
    
    # retrieve punctuation
    text = erase_punctuation(text)
    
    #lowercase
    text = text.lower()
    
    result = []
    for token in text.split(' '):
        if (token not in gensim.parsing.preprocessing.STOPWORDS) :
            if len(token)>3 : 
                token = lemmatize_stemming(token)
            if (token not in NEGATIVE_KEYWORDS):
                result.append(token)
    return result
```

```python
def flatten(lst):
    return [elem for sublist in list(map(preprocess2,lst)) for elem in sublist]
```

```python
Languages = flatten(Languages)
DBMS = flatten(DBMS)
Web = flatten(Web)
Roles = flatten(Roles)
```

### Rule based topic assigning

```python
def assign_general_and_refined_topic(text):
    activate_print = False
    
    if activate_print : print(text)
    
    def most_frequent(List): 
        if len(List)==0 : return 'Undetermined'
        return max(set(List), key = List.count) 

    
    keys = ['Programming Language', 'Web and Design', 'DB system']
    Topics = {keys[0]: Languages, keys[1] : Web, keys[2] : DBMS}
    prog, web, db = 0,0,0
    
    for token in text:
        if token in Languages: prog=prog+1
        if token in Web: web=web+1
        if token in DBMS: db=db+1
    
    values = [prog, web, db]
    idx = values.index(max(values))
    
    general = keys[idx]
    if activate_print : print('general ', general)
        
    refined = most_frequent([elem for elem in text if elem in Topics.get(general)])
    
    if activate_print : print('\n')
    
    return [general,refined]
```

```python
course_details['Topics'] = course_details['Preprocess'].apply(assign_general_and_refined_topic)
```

```python
course_details['General Topic'] = course_details['Topics'].apply(lambda x : x[0])
course_details['Refined Topic'] = course_details['Topics'].apply(lambda x : x[1])
course_details.loc[course_details['Refined Topic']=='Undetermined','General Topic'] = 'Undetermined'
course_details.drop('Topics',axis=1, inplace=True)
course_details.head(6)
```

Predicting missing Topics :

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
features = tfidf.fit_transform(df.title).toarray()
labels = df.category_id
features.shape
```

```python
from sklearn.feature_selection import chi2
import numpy as np
N = 2
for Product, category_id in sorted(category_to_id.items()):
    features_chi2 = chi2(features, labels == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    print("\n # '{}':".format(Product))
    print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-N:])))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(df['title'], df['General Topic'], random_state = 0)
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
clf = MultinomialNB().fit(X_train_tfidf, y_train)
```

```python
def predict_missing_gen_topic(text):
    return clf.predict(count_vect.transform(["Ruby Programming for Beginners"]))[0]
```

```python
course_details[course_details['General Topic']=='Undetermined']['title'].apply(predict_missing_gen_topic).value_counts()
```
